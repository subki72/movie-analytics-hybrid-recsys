```python
print(f"Shape sebelum menghapus duplikat: {movies_data.shape}")
movies_data = movies_data.drop_duplicates()
print(f"Shape setelah menghapus duplikat: {movies_data.shape}")
```

### **Handling Missing Value**

```python
movies_data.isna().sum()
```

```python
movies_data['release_date'].isna().sum()
```

```python
movies_data.dropna(subset=['release_date'], inplace=True)
```

```python
movies_data['release_date'].isna().sum()
```

```python
text_columns = movies_data.select_dtypes(include=['object']).columns
movies_data[text_columns] = movies_data[text_columns].fillna("No Data")
```

```python
median_runtime = movies_data['runtime'].median()
movies_data['runtime'] = movies_data['runtime'].fillna(median_runtime)
```

```python
movies_data = movies_data.fillna(0)
```

```python
movies_data.info()
```

```python
movies_data.isna().sum()
```

```python
released_movies = movies_data[movies_data['status'] == 'Released']
top_10_movies = released_movies.sort_values(by='vote_average', ascending=False).head(10)

print("=== 10 Films that have been released with the highest votes ===")
print(top_10_movies[['title', 'vote_average', 'vote_count', 'release_date']].to_string())
```

```python
runtime_movies = movies_data[movies_data['runtime'] > 180]
runtime_movies_sorted = runtime_movies.sort_values(by='runtime', ascending=False)

print("\n=== The film is over 3 hours long===")
print(runtime_movies_sorted[['title', 'runtime', 'release_date']].head(10))
```

```python
english_movies = movies_data[movies_data['original_language'] == 'en']
english_movies_sorted = english_movies.sort_values(by='popularity', ascending=False)

print("\n=== The film with english language===")
print(english_movies_sorted[['title', 'runtime', 'release_date', 'original_language']].head(10))
```

```python
long_english_released_movies = english_movies[english_movies['runtime'] > 180]
long_english_released_movies_sorted = long_english_released_movies.sort_values(by='runtime', ascending=False)

print("\n=== Top 10 English Released Films over 3 hours long ===")
print(long_english_released_movies_sorted[['title', 'runtime', 'release_date']].head(10).to_string())
```

### **Data Understanding & Preparation**

```python
rating.head(10)
```

```python
rating.info()
```

### **Aggregation & Merge**

```python
movies_rating = rating.groupby("movieId")["rating"].agg(["mean", "median", "count"])
```

```python
movies_rating
```

```python
english_movies_rating = pd.merge(english_movies, movies_rating, left_on="id", right_on="movieId", how="inner")
```

### **Create a profit column = revenue - budget**

```python
english_movies_rating ["profit"] = english_movies_rating ["revenue"] - english_movies_rating ["budget"]
```

```python
english_movies_rating[['original_title', 'profit', 'runtime']].sort_values(by='profit', ascending=False).head(10)
```

### **Create a language column to determine whether it's English or not (is_English)**

```python
released_movies_english = released_movies['original_language'] == 'en'
```

```python
released_movies_english.head()
```

### **Create a movie title length column (title_len)**

```python
movies_data['title_len'] = movies_data['title'].apply(len)
display(movies_data[['title', 'title_len']].head())
```

### **Create a runtime column from minutes to hours (0-n) (duration_hours)**

```python
movies_data['duration_hours'] = movies_data['runtime'] / 60
display(movies_data[['title', 'runtime', 'duration_hours']].head())
```

### **Create a favorite movies column with a median rating >= 3**

```python
english_movies_rating['favorite_movies'] = english_movies_rating['median'] >= 3
display(english_movies_rating[['title', 'median', 'favorite_movies']].head())
```

# **Data Exploration**

```python
english_movies_rating["favorite_movies"] = (english_movies_rating["median"] >= 3).astype(int)
english_movies_rating["favorite_movies"].value_counts().plot(kind="bar")
plt.title("Favorite Movies English Distribution")
plt.show()
```

```python
movies_data['duration_hours'] = movies_data["runtime"].div(60).round(0).astype("int")
movies_data['duration_hours'].value_counts().sort_index().plot(kind="bar")
plt.title("Movie Duration Distribution")
plt.show()
```
